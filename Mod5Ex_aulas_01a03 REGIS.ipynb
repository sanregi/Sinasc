```python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
# 3) seu código aqui
sinasc_dados = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO','GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']]
```

```python
# 4) seu código aqui
sinasc_dados['APGAR5'].isna().sum()
```

```python
# 5) seu código aqui
sinasc_dados.loc[:, ['ESTCIVMAE', 'CONSULTAS']].fillna(value={'ESTCIVMAE': 9, 'CONSULTAS': 9}, inplace=True)

sinasc_dados
```

```python
# 6) Seu código aqui

sinasc_dados.loc[:, ['QTDFILVIVO']].fillna(value=0, inplace=True)
```

```python
# 7) seu código aqui

sinasc_dados.loc[:, ['ESCMAE']].fillna(value=9, inplace=True)
sinasc_dados.loc[:, ['GESTACAO']].fillna(value=9, inplace=True)
sinasc_dados.loc[:, ['GRAVIDEZ']].fillna(value=9, inplace=True)

sinasc_dados
```

De acordo com o dicionário de dados que está no arquivo estrutura_sinasc_para_CD.pdf, as três últimas variáveis com dados faltantes 'ESCMAE', 'GESTACAO' e 'GRAVIDEZ', possuem o código 9 que significa ignorado. 

```python
# 8) seu código aqui
sinasc_dados.loc[sinasc_dados['APGAR5'] >= 8, "ASFIXIA"] = 'Normal'
sinasc_dados.loc[(sinasc_dados['APGAR5'] >= 6) & (sinasc_dados['APGAR5'] <=7), "ASFIXIA"] = 'Leve'
sinasc_dados.loc[(sinasc_dados['APGAR5'] >= 4) & (sinasc_dados['APGAR5'] <=5), "ASFIXIA"] = 'Moderada'
sinasc_dados.loc[sinasc_dados['APGAR5'] <=3 , "ASFIXIA"] = 'Severa'

sinasc_dados
```

```python
sinasc_dados['ASFIXIA'].value_counts()
```

```python
# 9) seu código aqui
sinasc_dados.rename(columns={'LOCNASC': 'local_de_nascimento',
                                'IDADEMAE': 'idade_da_mae',
                                'ESTCIVMAE': 'estado_civil_da_mae',
                                'ESCMAE': 'escolaridade_da_mae__anos_de_estudo_concluídos',
                                'QTDFILVIVO': 'quantidade_de_filhos_vivos',
                                'GESTACAO': 'semanas_de_gestacao',
                                'GRAVIDEZ': 'tipo_de_gravidez',
                                'CONSULTAS': 'numero_de_consultas_de_prenatal',
                                'APGAR5': 'apgar_no_quinto_minuto'},
                                inplace=True)
sinasc_dados
```
